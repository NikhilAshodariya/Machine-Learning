```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from sklearn.cross_validation import train_test_split
```

```python
def toList(df):
    idx =df.index.tolist()
    columns=df.columns
    l=[]
    for i in idx:
        temp=[]
        for j in columns:
            a=df.get_value(i,j)
            temp.append(a)
        l.append(temp)
    return l
```

```python
dataFrame=pd.read_csv('clustering.csv')
```

```python
features=dataFrame.columns[:-1]
target=dataFrame.columns[-1]
```

```python
train,test=train_test_split(dataFrame)
```

```python
lis=toList(train)
```

```python
colors = ["g.","r.","c.","y.","k.",'-c.','r.','g.']
```

```python
kmeans6 = KMeans(n_clusters=6)
kmeans6.fit(lis)

centroids = kmeans6.cluster_centers_
labels = kmeans6.labels_

print(centroids)
print(labels)

for i in range(len(lis)):
#     print("coordinate:",lis[i], "label:", labels[i])
    plt.plot(lis[i][0], lis[i][1], colors[labels[i]], markersize = 10)
# plt.show()

plt.scatter(centroids[:, 0],centroids[:, 1], marker = "x", s=150, linewidths = 5, zorder = 10)

plt.show()
```
